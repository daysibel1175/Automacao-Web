```python
# passo 1: pegar a cotaçao do dolar 
#  -abrir navegador - entrar en google - pesquisar a cotaçao do dolar no google - pegar a cotaçao do dolar do google
# para pegar o xpath = inspeccionar + selector + selecionar a caixa de texto do google + boton direito encima do texto em azul + copy > copy Xpath
# .click para clickar, .get get_attribute para pegar info, .send_keys para escrever 
from selenium import webdriver #permite criar um navegador
from selenium.webdriver.common.keys import Keys #permite escrever no navegador
from selenium.webdriver.common.by import By #permite selecionar itens no navegador

navegador = webdriver.Chrome()
navegador .get("https://google.com/")
navegador.find_element("xpath", '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotaçao dolar")
navegador.find_element("xpath", '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

cotacao_dolar = navegador.find_element("xpath",'//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute("data-value")
print(cotacao_dolar)

# passo 2: pegar a cotaçao do euro

navegador = webdriver.Chrome()
navegador .get("https://google.com/")
navegador.find_element("xpath", '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotaçao euro")
navegador.find_element("xpath", '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

cotacao_euro = navegador.find_element("xpath",'//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute("data-value")
print(cotacao_euro)
                                                                                              
# passo 3: pegar a cotaçao do ouro
#como python só reconhece as casas decimais por o ponto, os nros com virgula precisamos usar replace = para mudar , por .
#como as cotaçoes vem de um site o python le como texto, tambem precisamos mudar isso

navegador.get("https://www.melhorcambio.com/ouro-hoje")
cotacao_ouro = navegador.find_element("xpath", '//*[@id="comercial"]').get_attribute("value")
cotacao_ouro = cotacao_ouro.replace("," , ".")
print(cotacao_ouro)

navegador.quit() #para fechar o navegador
```

```python
# passo 4: atualizar a base dos preços (atualizando o preço de venda e o de compra)

import pandas as pd

tabela = pd.read_excel("Produtos.xlsx")
display(tabela)

# tabela.loc[nro da linha, nome da coluna] = float ou int(cotacao_dolar) seria a forma de pegar uma linha de uma coluna(loc de localizar)

tabela.loc[tabela["Moeda"] == "Dólar", "Cotação"] = float(cotacao_dolar)

tabela.loc[tabela["Moeda"] == "Euro", "Cotação"] = float(cotacao_euro)

tabela.loc[tabela["Moeda"] == "Ouro", "Cotação"] = float(cotacao_ouro)

display(tabela)
```

```python
# tabela.loc[nro da linha, nome da coluna] = float ou int(cotacao_dolar) seria a forma de pegar uma linha de uma coluna(loc de localizar)
#tabela.loc[tabela["Moeda"] == "Dólar", : ] = float(cotacao_dolar) esses dos pontos quer dizer todas as colunas
# .map("R${:.2f}.format") codigo para formatar os nros
# -atualizar: - coluna de cotaçao 

tabela.loc[tabela["Moeda"] == "Dólar", "Cotação"] = float(cotacao_dolar)

tabela.loc[tabela["Moeda"] == "Euro", "Cotação"] = float(cotacao_euro)

tabela.loc[tabela["Moeda"] == "Ouro", "Cotação"] = float(cotacao_ouro)

# - coluna preço de compra ( preço original * cotaçao ) 

tabela["Preço de Compra"] = tabela["Preço Original"] * tabela["Cotação"]

# - coluna preço de venda ( preço de compra * margem )

tabela["Preço de Venda"] = tabela["Preço de Compra"] * tabela["Margem"]

display(tabela)
```
